# 瞬発力を高める直前確認項目

大学受験数学は暗記科目と言われており,
公式・解法を暗記すれば攻略可能と言われています。
一方, 試験本番中に

「あれ, 公式これで正しかったかな...?」

と動揺して時間をロスしさらに動揺して時間をロスし,
という悪循環に陥ることがあるかも知れません(少なくとも私はそうです)。
この確認項目集には数IIIに関係が深いものを中心に,
頑張れば暗算できる程度に簡単な計算で公式・解法が間違って無さそうなことを確認する方法をまとめました。

紙に書いて計算しても良いのですが, どちらかというと通学・移動時間,
歯磨きの時間などの隙間時間を利用して暗算・イメージで確認するのが効果的です。
- 何も見ないで簡単な計算や図をイメージできれば答案を書くスピードが数倍になります。
- 問題集等の解説を読むスピードも上がります。
- 確認手段自体が問題を解く方法を含みます。
- いつもと違って意図的に図を載せていません
  自分の頭でイメージしてみましょう。
- 他の科目や塾の宿題, 過去問等もあると思うので,
  最初は 1 日 10 分程度で良いと思います。
  その上でこの学習法が自身に合うか合わないか考えて時間配分を考えれば良いでしょう。

最初はつらいです。だんだん慣れてきます。気合を入れすぎず, ダラダラ何周もしましょう。
- 1 周目は読みすらしない, 視界に入れさえすればよい。
- 2 周目で文章を目読する。
- 3 周目でどんな問題か理解しようとする。理解できなくてもよい。
- 4 周目でわかるものから考える。わからないものは飛ばしてよい。
  - いや, 4 周目くらいでわからないものを考えこむのは時間の無駄なことが多いので,
    恐れずに飛ばさなければならない。
- 5, 6 周目も同じ。だんだんわかる問題が増えてくる。
- わからない問題を調べつつ何周もする。
- 10 周目くらいでもちんぷんかんぷんな問題があれば教えてください。
  今後の改善材料に致したく。また, 質問受け付けます。
- 慣れてきたら公式の箇所を黒く塗りつぶす等して隠しても良い。
- 本命試験 1 週間前に 1 日 1 周できるようになるのが理想的。

## 数と式, 2次関数
---

1. $y=(x-1)^2$ について
    1. $x=-1, 0, 1, 2$ を代入し, $y=4, 1, 0, 1$ となることを確認せよ.
    1. それが $y=x^2$を右に$1$移動したグラフになることを確認せよ。
1. 2次多項式 $-3x^2+30x$ が
   $=-3(x^2-10x)$ $=-3(x^2-2\cdot5x+5^2-5^2)$
   $=-3(x-5)^2+3\cdot5^2$

1. 円の方程式 $(x-1)^2+y^2=5^2$ について

       となることを確認せよ。
    1. その円が $x^2+y^2=5^2$ を右に 1 移動したものであることを確認せよ。
1. 楕円 $\displaystyle{x^2+\left(\frac{y}{2}\right)^2=5^2}$ について
    1. $x=0, 3, 4, 5$ を代入し $y=\pm10, \pm8, \pm6, 0$
       となることを確認せよ。
    1. その楕円が円 $x^2+y^2=5^2$ を上下に 2 倍に引き延ばした形になることを確認せよ。

## 数列
---

1. 数列の和に関する確認である。
    1. $1+4+7+10$ を考える時これと $10+7+4+1$
       と加算することにより $11+11+11+11=11\cdot4$ となり,
       これを $2$ で割ると $10+7+4+1$ と一致することを確認せよ。
    1. これを拡張して一般に等差数列の和の公式が得られることを確認せよ。
    1. $S=1+r+r^2+r^3$ とする。$S-rS=1-r^4$ となることを確認せよ。
    1. 同様に $S=1+r+r^2+\cdots+r^{n-2}+r^{n-1}$ とした時に
       $S-rS=1-r^n$ が得られることを確認せよ。
    1. 上の式を $S$ について解き,
       初項が $1$ の時の等比数列の和の公式が導かれることを確認せよ。

1. $\Sigma$ の公式に関する確認である。
    1. $\displaystyle\sum_{k=1}^nk=\frac{1}{2}n(n+1)$
       の両辺に $3$ を代入し, 両辺の計算結果が一致することを確認せよ。
    1. $\displaystyle\sum_{k=1}^nk^2=\frac{1}{6}n(n+1)(2n+1),$
       $\displaystyle\sum_{k=1}^nk^3=\frac{1}{4}\{n(n+1)\}^2$
       に対して同様の確認をせよ。

1. 2 項定理 $\displaystyle(a+b)^n=\sum_{r=0}^{n}{}_nC_ra^{n-r}b^r$
   に関する確認である。
    1. $n=0$ の時, 両辺共に $1$ となることを確認せよ。
    1. $n=1$ の時, 両辺共に $(a+b)$ となることを確認せよ。
    1. $n=2$ の時, 左辺は $(a+b)^2$, 右辺は
       ${}_2C_0a^2b^0+{}_2C_1a^1b^1+{}_2C_2a^0b^2$
       となる。これらを整理し両辺一致することを確認せよ。
    1. $n=3$ の時, 同様のことを確認せよ。
    1. パスカルの三角形がなんであるか調べ, 成分が 2 項係数
       ${}_nC_r$ に一致することを確認せよ。
    1. $a=1, b=0.01$ として $1.01^2=1.0201$, $1.01^3=1.030301$,
       $1.01^4=1.04060401$ となることを確認せよ。
    1. 2 項定理は数学的帰納法により証明できる。
       余力があれば数学的帰納法自体の練習になるのでやってみるとよい。

## 指数・対数関数
---

1. 指数法則についての確認である。
    1. $2^2\cdot2^3$ $=4\cdot8$ $=32=2^5=2^{2+3}$
       であることを確認せよ。
    1. より一般に実数 $b$(底:base の頭文字)と自然数 $m, n$ に対して
       $b^m\cdot b^n=b^{m+n}$ となることを確認せよ。
    1. $b^0\cdot b^n=b^{0+n}=b^n$ と考えて $b^0$
       を掛け算しても値が変化しないように $b^0=1$
       と約束するのが便利であることを確認せよ。
    1. $b^n\cdot b^{-n}=b^{n-n}=b^0=1$
       と考えて $b^{-n}$ と $b^n$ の積が 1 となるように
       $\displaystyle{b^{-n}=\frac{1}{b^n}}$
       と約束するのが便利であることを確認せよ。
    1. $(2^2)^3=4^3=64=2^6=2^{2\cdot3}$ であることを確認せよ。
    1. より一般に実数 $b$ と自然数 $m, n$ に対して
       $(b^m)^n=b^{mn}$ となることを確認せよ。
    1. $({b^\frac{1}{2}})^2=b^{\frac{1}{2}\cdot 2}=b^{1}=b$
       と考えて $b^\frac{1}{2}$ を 2 乗すると $b$ になるように
       $b^\frac{1}{2}=\sqrt{b}$
       と約束するのが便利であることを確認せよ。
    1. 同様に $b^\frac{1}{n}=\sqrt[n]{b}$ ($b$ の実数 $n$ 乗根)
       と約束するのが便利であることを確認せよ。

1. 底の変換についての確認である。
    1. $x = b^y$ の時 $y = \log_b x$ であることを確認せよ。(対数の定義)
    1. $x = b^y$ の両辺に別の底 $c$ の対数を採り,
       底の変換公式 $\log_c x = (\log_b x) (\log_c b)$
       を導け。

1. 対数法則についての確認である。
    1. 指数関数と対数関数がお互い逆関数であることを考え
       $x=b^{\log_b x},$ $w=b^{\log_b w}$
       であることを確認せよ。
    1. $b^{\log_b xw}$
       $=xw$
       $=b^{\log_b x}b^{\log_b w}$
       $=b^{\log_b x+\log_b w}$
       となることを確認せよ。
    1. 上の式の対数を採ることにより
       $\log_b xw = \log_b x + \log_b w$
       となることを確認せよ。

## ベクトル
---

1. 直線 $l:\,ax+by+c=0$ について $l$ と $x$, $y$
   軸との交点をそれぞれ $A$, $B$ とする。原点を $O$ とする。
    1. $l$ に $y=0$ を代入し, $A$ の $x$ 座標が
       $\displaystyle{-\frac{c}{a}}$ であることを確認せよ。
    1. $l$ に $x=0$ を代入し, $B$ の $y$ 座標が
       $\displaystyle{-\frac{c}{b}}$ であることを確認せよ。
    1. $\triangle OAB$ の面積が
       $\displaystyle\frac{c^2}{2|ab|}$ であることを確認せよ。
    1. $AB$ の長さが
       $\displaystyle{\sqrt{\left(\frac{c}{a}\right)^2+\left(\frac{c}{b}\right)^2}}$
       $\displaystyle{=\sqrt{\frac{c^2b^2+c^2a^2}{a^2b^2}}}$
       $\displaystyle{=\left|\frac{c}{ab}\right|\sqrt{a^2+b^2}}$
       となることを確認せよ。
    1. $AB$ を底辺としたときの $\triangle OAB$ の高さが
       $\displaystyle\frac{2\triangle OAB}{AB}$
       $\displaystyle=\frac{c^2}{|ab|}\frac{|ab|}{|c|\sqrt{a^2+b^2}}$
       $\displaystyle=\frac{|c|}{\sqrt{a^2+b^2}}$
       となることを確認せよ。
    1. これが原点 $O$ と直線 $l$ との距離であることを確認せよ。
    1. 直線 $l$ を $x$ 方向に $-x_1$, $y$ 方向に $-y_1$
       移動すると, 移動後の直線の方程式が
       $a(x+x_1)+b(y+y_1)+c$ $=ax+by+ax_1+by_1+c=0$
       となることを確認せよ。
    1. 点 $(x_1, y_1)$ と直線 $ax+by+c=0$ の距離が
       $\displaystyle\frac{|ax_1+by_1+c|}{\sqrt{a^2+b^2}}$
       となることを確認せよ。

1. 点 $A$ の座標を $(2, 0)$, 点 $B$ の座標を $(0, 2)$ とする。
   原点を $O$ とする。この時, 直線 $AB$ の方程式は媒介変数 $t$
   を用いて
   $(1-t)\overrightarrow{OA}+t\overrightarrow{OB}$
   $=(1-t)(2, 0) + t(0, 2)$
   と表される。
    1. $t=0$ を代入すると点 $A$ の座標を得られることを確認せよ。
    1. $t=1$ を代入すると点 $B$ の座標を得られることを確認せよ。
    1. $\displaystyle t=\frac{1}{2}$ を代入すると線分 $AB$
       の中点の座標を得られることを確認せよ。
    1. 直線 $AB$ の $x, y$ 座標がそれぞれ
       $x=2-2t$, $y=2t$ となることを確認せよ。
    1. $t$ を消去して直線 $AB$ の方程式
       $x+y-2=0$ が得られることを確認せよ。
    1. 原点と直線の距離の公式
       $\displaystyle\frac{|c|}{\sqrt{a^2+b^2}}$
       を使用し, 原点 $O$ と直線 $AB$
       の距離 $\sqrt{2}$ が得られることを確認せよ。
    1. 点と直線の距離の公式
       $\displaystyle\frac{|ax_1+by_1+c|}{\sqrt{a^2+b^2}}$
       を使用し, 点 $(2, 2)$ と直線 $AB$ の距離が
       $\sqrt{2}$ が得られることを確認せよ。

## 複素数
---

1. 複素数の積と商についての確認である。以下実数 $a, b, c, d$
   を用いて $z=a+ib, w=c+id$ とする。
    1. $zw$ $=(a+ib)(c+id)$ $=ac-bd+i(ad+bc)$ となることを確認せよ。
    1. $\overline{z}\overline{w}=(a-ib)(c-id)$ が
       $\overline{zw}=ac-bd-i(ad+bc)$ と一致することを確認せよ。
    1. $\displaystyle\frac{z}{w}=\frac{a+ib}{c+id}$ の分母と分子に
       $c-id$ を乗算することで
       $\displaystyle\frac{ac+bd+i(-ad+bc)}{c^2+d^2}$
       となることを確認せよ。
    1. $\displaystyle\frac{\overline{z}}{\overline{w}}=\frac{a-ib}{c-id}$
       が
       $\displaystyle\overline{\left(\frac{z}{w}\right)}$
       $\displaystyle=\frac{ac+bd+i(ad-bc)}{c^2+d^2}$
       と一致することを確認せよ。そのまま計算するのは大変なので,
       $\displaystyle\frac{z}{w}$ の $b$ を $-b$, $d$ を $-d$
       に置き換えたものを考えればよい。

1. $z=\cos\alpha+i\sin\alpha$, $w=\cos\beta+i\sin\beta$ とする。
    1. $zw=(\cos\alpha+i\sin\alpha)(\cos\beta+i\sin\beta)$
       を展開し三角関数の加法定理
       $\cos(\alpha+\beta)=\cos\alpha\cos\beta-\sin\alpha\sin\beta$,
       $\sin(\alpha+\beta)=\sin\alpha\cos\beta+\cos\alpha\sin\beta$
       を適用し
       $zw=\cos(\alpha+\beta)+i\sin(\alpha+\beta)$
       となることを確認せよ。
    1. $z^n=\cos n\alpha+i\sin n\alpha$ となることを確認せよ。
    1. $1$ の複素 $n$ 乗根が $0\leqq k<n$ となる整数 $k$ を用いて
       $\displaystyle\cos\frac{2\pi k}{n}+\sin\frac{2\pi k}{n}$
       と表されることを確認せよ。

## 微分
---

1. 関数 $\displaystyle{f(x)=\begin{cases}x^2&(x\neq 0)\\1&(x=0)\end{cases}}$ について
    1. $\displaystyle{\lim_{x\rightarrow0}f(x)=0}$ となることを確認せよ。
    1. $\displaystyle{\lim_{x\rightarrow0}f(x)\neq f(0)}$ であることから $f(x)$ は
       $x=0$ において非連続であることを確認せよ。

1. 関数 $f(x)=|x|$ について
    1. $x>0$ における $f'(x)$ を求めよ。
    1. $x<0$ における $f'(x)$ を求めよ。
    1. $\displaystyle{\lim_{x\rightarrow-0}f'(x)\neq\lim_{x\rightarrow+0}f'(x)}$
       であることから $f(x)$ は $x=0$ で微分可能でないことを確認せよ。

1. $f(x)=x^2$ について
    1. $x=1$, $h=0.01$ として
       $\displaystyle{\frac{f(x+h)-x(x)}{h}}$
       を計算せよ。分子は $0.0201$ になる。
    1. $h=0.001$, $h=0.0001$ とした時に
       $\displaystyle{\frac{f(x+h)-x(x)}{h}}$ が $2$
       に近づくことを確認せよ。

1. 積の微分公式 $\{f(x)\,g(x)\}'=f'(x)\,g(x)+f(x)\,g'(x)$ について
    1. $f(x)=x$, $g(x)=x$ として左辺 $(x^2)'$ と右辺 $x'x + xx'$
       を計算・整理し, 一致することを確認せよ。
    1. $f(x)=x$, $g(x)=x^2$ として左辺と右辺を計算し, 一致することを確認せよ。

1. 合成関数の微分公式 $\{f(g(x))\}'=f'(g(x))\,g'(x)$ について
   $f(y)=y^{-1}$, $g(x)=x^2$ として
    1. 左辺が $(x^{-2})'$ となることを確認せよ。
    1. 右辺が $\{(x^2)^{-1}\}'=-(x^2)^{-2}\cdot 2x$ となることを確認せよ。
    1. 左辺と右辺の計算結果を整理し, 一致することを確認せよ。

1. 逆関数の微分公式
   $\displaystyle{\{f^{-1}(x)\}'=\frac{1}{f'(x)}}$
   について
   $x\leqq 0$ を考え $f(x) = x^2$ とする。
    1. $f^{-1}(x)=\sqrt{x}$ となることを確認せよ。
    1. $\{\sqrt{x}\}'$ を計算し,
       $\displaystyle{\frac{1}{f'(x)}}$
       と一致することを確認せよ。

1. 陰関数表示 $y^2=x^3$ について
    1. 陽関数表示 $y=\pm x^{\frac{3}{2}}$ として
       $\displaystyle{y'=\pm\frac{3}{2}x^{\frac{1}{2}}}$
       となることを確認せよ。
    1. 陰関数表示のまま両辺を微分すると $2yy'=3x^2$ となることを確認せよ。
    1. それを $y'$ について解くと
       $\displaystyle{y'=\frac{3x^2}{2y}}$
       となることを確認せよ。
    1. それに $y=\pm x^{\frac{3}{2}}$ を代入し,
       陽関数表示を微分した場合と一致することを確認せよ。

1. 三角関数 $\sin x$ の微分についての確認である。
    1. 微分の定義
       $\displaystyle{\lim_{h\rightarrow0}\frac{f(x+h)-f(x)}{h}}$
       により
       $\displaystyle{(\sin x)'=\lim_{h\rightarrow0}\frac{\sin(x+h)-\sin x}{h}}$
       となることを確認せよ。
    1. $\sin(x+h)=\sin x\cos h+\cos x\sin h$ と
       $\displaystyle{\lim_{x\rightarrow0}\frac{\sin x}{x}=1}$
       を考えて $(\sin x)'=\cos x$ となることを確認せよ。
    1. 単位円上での $\displaystyle{\frac{\pi}{2}}$ 回転を考え
       $\displaystyle{(\cos x)'}$
       $\displaystyle{=\{\sin(x+\frac{\pi}{2})\}'}$
       $\displaystyle{=\cos(x+\frac{\pi}{2})}$
       $\displaystyle{=-\sin(x)}$
       となることを確認せよ。

1. 対数関数 $\log_b x$ の微分についての確認である。
    1. 微分の定義により
       $\displaystyle{(\log_b x)'=\lim_{h\rightarrow0}\frac{\log_b(x+h)-\log_b x}{h}}$
       となることを確認せよ。
    1. 対数の差が真数の比に対応することから上の式が
       $\displaystyle{\lim_{h\rightarrow0}\frac{1}{h}\log_b\frac{x+h}{x}}$
       $=\displaystyle{\lim_{h\rightarrow0}\frac{1}{h}\log_b\left(1+\frac{h}{x}\right)}$
       となることを確認せよ。
    1. それをさらに
       $\displaystyle{\lim_{h\rightarrow0}\frac{1}{h}\frac{x}{x}\log_b\left(1+\frac{h}{x}\right)}$
       $=\displaystyle{\frac{1}{x}\lim_{h\rightarrow0}\frac{x}{h}\log_b\left(1+\frac{h}{x}\right)}$
       $=\displaystyle{\frac{1}{x}\lim_{h\rightarrow0}\log_b\left(1+\frac{h}{x}\right)^\frac{x}{h}}$
       $=\displaystyle{\frac{1}{x}\log_b\lim_{h\rightarrow0}\left(1+\frac{h}{x}\right)^\frac{x}{h}}$
       と変形できることを確認せよ。
    1. $h\rightarrow0$ の時 $\displaystyle\frac{h}{x}\rightarrow0$
       なので上の式の $\displaystyle\frac{h}{x}$ を $h$ に置き換えて
       $=\displaystyle{\frac{1}{x}\log_b\lim_{h\rightarrow0}(1+h)^\frac{1}{h}}$
       $=\displaystyle\frac{1}{x}\log_b e$
       と変形できることを確認せよ。
    1. 特に $b=e$ の時 $\displaystyle(\log x)'=\frac{1}{x}$
       となることを確認せよ。
