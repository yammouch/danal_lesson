# 瞬発力を高める直前確認項目

大学受験数学は暗記科目と言われており,
公式・解法を暗記すれば攻略可能と言われています。
一方, 試験本番中に

「あれ, 公式これで正しかったかな...?」

と動揺して時間をロスしさらに動揺して時間をロスし,
という悪循環に陥ることがあります(少なくとも私はそうです)。
この確認項目集には数IIIに関係が深いものを中心に,
頑張れば暗算できる程度に簡単な計算で公式・解法が間違って無さそうなことを確認する方法をまとめました。

- 紙に書いて計算しても良いのですが, どちらかというと通学・移動時間,
  歯磨きの時間などの隙間時間を利用して暗算・イメージで確認するのが効果的です。
  - 何も見ないで簡単な計算や図をイメージできればスピードが数倍になります。
  - 試験本番では念のため書いて確認してもよいですが,
    速く書けるようになります。
  - いつもと違って意図的に図を載せていません
    自分の頭でイメージしてみましょう。
  - 他の科目や塾の宿題等もあると思うので,
    最初は 1 日 10 分程度で良いと思います。
    その上でこの学習法が自身に合うか合わないか考えて配分を考えれば良いでしょう。
- 最初はつらいです。だんだん慣れてきます。気合を入れすぎず, ダラダラ何周もしましょう。
  - 1 周目は読みすらしない, 視界に入れさえすればよい。
  - 2 周目で文章を目読する。
  - 3 周目でどんな問題か理解しようとする。理解できなくてもよい。
  - 4 周目でわかるものから考える。わからないものは飛ばしてよい。
  - 5, 6 周目も同じ。だんだんわかる問題が増えてくる。
  - わからない問題を調べつつ何周もする。
  - 慣れてきたら公式の箇所を黒く塗りつぶす等して隠しても良い。
  - 本命試験 1 週間前に 1 日 1 周できるようになるのが理想的。

## 数と式, 2次関数
---

1. $y=(x-1)^2$ に $x=-1, 0, 1, 2$ を代入し, $y=4, 1, 0, 1$ となること, またそれが $y=x^2$を右に$1$移動したグラフになることを確認せよ。
1. 2次多項式 $3x^2+30x$ が $=3(x^2+10x)$ $=3(x^2+2\cdot5x+5^2-5^2)$ $=3(x+5)^2-3\cdot5^2$ と平方完成できることを確認せよ。
1. 円の方程式 $(x-1)^2+y^2=5^2$ に $x=1, 4, 5, 6$ を代入して $y=\pm5, \pm4, \pm3, 0$ となること, またその円が $x^2+y^2=5^2$ を右に 1 移動したものであることを確認せよ。
1. 楕円 $\displaystyle{x^2+\left(\frac{y}{2}\right)^2=5^2}$ に $x=0, 3, 4, 5$ を代入し $y=\pm10, \pm8, \pm6, 0$ となること, またその楕円が円 $x^2+y^2=5^2$ を上下に 2 倍に引き延ばした形になることを確認せよ。

## 三角関数
---

1. 原点を中心とする半径 1 の円(単位円)において
    1. $\sin\theta, \cos\theta, \tan\theta$
       がどのように表せられるか確認せよ。
    1. $\sin^2+\cos^2=1,$
       $\displaystyle{\tan\theta=\frac{\sin\theta}{\cos\theta}}$
       となることを確認せよ。
    1. $\sin^2\theta+\cos^2\theta=1$ の両辺を $\cos^2\theta$ で割り,
       $\displaystyle{1+\tan^2\theta=\frac{1}{\cos^2\theta}}$
       となることを確認せよ。

1. 直角三角形とその外接円を考え, 正弦定理
   $\displaystyle{\frac{a}{\sin A}=\frac{b}{\sin B}=\frac{c}{\sin C}=2R}$
   が成り立つことを確認せよ。

1. $\triangle ABC$ の頂点 $C$ から辺 $AB$ に垂線を垂らし,
   その足を $D$ とする。$a=BC, b=CA, c=AB$ とする。
    1. $AD=b\cos A, CD=b\sin A, BD=c-b\cos A$
       となることを確認せよ。
    1. $AB$ と底辺とした場合の高さは $CD$ になることから,
       $\triangle ABC$ の面積が $\displaystyle{\frac{1}{2}bc\sin A}$
       となることを確認せよ。
    1. 直角三角形 $BCD$ に三平方の定理を適用すると
       $a^2=(b\sin A)^2+(c-b\cos A)^2$ となること,
       またそれを展開し余弦定理が成り立つことを確認せよ。

1. 以下の三角関数の周期を確認せよ。
    1. $\sin\theta$ の周期が $2\pi$ であること。
    1. $\sin2\theta$ の周期が $\pi$ であること。
    1. $\displaystyle{\sin\frac{\theta}{2}}$
       の周期が $4\pi$ であること。

1. 以下の三角関数の積を求めるために,
   $\sin(\alpha\pm\beta), \cos(\alpha\pm\beta)$
   のどの組み合わせの和をとるか差をとるか確認せよ。
    1. $\sin\alpha\cos\beta$
    1. $\cos\alpha\sin\beta$
    1. $\cos\alpha\cos\beta$
    1. $\sin\alpha\sin\beta$
    
1. $\displaystyle{\frac{A+B}{2}}$ と $\displaystyle{\frac{A-B}{2}}$
   について
    1. 和をとると $A$ になることを確認せよ。
    1. 差をとると $B$ になることを確認せよ。

1. 以下のそれぞれの三角関数の積が
   $\sin A$, $\sin B$, $\cos A$, $\cos B$
   のいずれか 2 つの組み合わせの和または差で表せることを確認せよ。
    1. $\displaystyle{\sin\frac{A+B}{2}\cos\frac{A-B}{2}}$
    1. $\displaystyle{\cos\frac{A+B}{2}\sin\frac{A-B}{2}}$
    1. $\displaystyle{\cos\frac{A+B}{2}\cos\frac{A-B}{2}}$
    1. $\displaystyle{\sin\frac{A+B}{2}\sin\frac{A-B}{2}}$

## 極限
---

1. $\sqrt{x+1}-\sqrt{x-1}$ について
    1. 分子を有理化せよ。
    1. 有理化後の式について $x\rightarrow 0$ の時分子は収束し,
       分母が発散することを確認せよ。

1. $x$ 軸との角度が $\theta$ で原点 $O$
   を通り第 1 象限に延びる半直線を $l$ とする。
   $l$ と単位円との交点を $A$, $l$ と直線 $x=1$ との交点を $B$,
   $A$ から $x$ 軸に垂らした垂線の足を $C$, 点 $(1, 0)$ を $D$ とする。
    1. $\triangle OAC$ の面積が
       $\displaystyle{\frac{1}{2}\sin\theta}$
       であることを確認せよ。
    1. 扇形 $OAD$ の面積が $\displaystyle{\frac{1}{2}\theta}$
       であることを確認せよ。
    1. $\triangle OBD$ の面積が
       $\displaystyle{\frac{1}{2}\tan\theta}$
       であることを確認せよ。
    1. $\triangle OAC$, 扇形 $OAD$, $\triangle OBD$
       の面積の大小関係から $\sin\theta < \theta < \tan\theta$
       が成り立つことを確認せよ。
    1. その不等式を $\sin\theta$ で割り, はさみうちの原理を考えて
       $\displaystyle{\lim_{\theta\rightarrow 0}\frac{\sin\theta}{\theta} = 1}$
       を確認せよ。

## 微分
---

1. 関数 $\displaystyle{f(x)=\begin{cases}x^2&(x\neq 0)\\1&(x=0)\end{cases}}$ について
    1. $\displaystyle{\lim_{x\rightarrow0}f(x)=0}$ となることを確認せよ。
    1. $\displaystyle{\lim_{x\rightarrow0}f(x)\neq f(0)}$ であることから $f(x)$ は
       $x=0$ において非連続であることを確認せよ。

1. 関数 $f(x)=|x|$ について
    1. $x>0$ における $f'(x)$ を求めよ。
    1. $x<0$ における $f'(x)$ を求めよ。
    1. $\displaystyle{\lim_{x\rightarrow-0}f'(x)\neq\lim_{x\rightarrow+0}f'(x)}$
       であることから $f(x)$ は $x=0$ で微分可能でないことを確認せよ。

1. $f(x)=x^2$ について
    1. $x=1$, $h=0.01$ として
       $\displaystyle{\frac{f(x+h)-x(x)}{h}}$
       を計算せよ。分子は $0.0201$ になる。
    1. $h=0.001$, $h=0.0001$ とした時に
       $\displaystyle{\frac{f(x+h)-x(x)}{h}}$ が $2$
       に近づくことを確認せよ。

1. 積の微分公式 $\{f(x)\,g(x)\}'=f'(x)\,g(x)+f(x)\,g'(x)$ について
    1. $f(x)=x$, $g(x)=x$ として右辺 $(x^2)'$ と左辺 $x'x + xx'$
       を計算・整理し, 一致することを確認せよ。
    1. $f(x)=x$, $g(x)=x^2$ として右辺と左辺を計算し, 一致することを確認せよ。

1. 合成関数の微分公式 $\{f(g(x))\}'=f'(g(x))\,g'(x)$ について
   $f(y)=y^{-1}$, $g(x)=x^2$ として
    1. 左辺が $(x^{-2})'$ となることを確認せよ。
    1. 右辺が $\{(x^2)^{-1}\}'=-(x^2)^{-2}\cdot 2x$ となることを確認せよ。
    1. 左辺と右辺の計算結果を整理し, 一致することを確認せよ。

1. 逆関数の微分公式
   $\displaystyle{\{f^{-1}(x)\}'=\frac{1}{f'(x)}}$
   について
   $x\leqq 0$ を考え $f(x) = x^2$ とする。
    1. $f^{-1}(x)=\sqrt{x}$ となることを確認せよ。
    1. $\{\sqrt{x}\}'$ を計算し,
       $\displaystyle{\frac{1}{f'(x)}}$
       と一致することを確認せよ。

1. 陰関数表示 $y^2=x^3$ について
    1. 陽関数表示 $y=\pm x^{\frac{3}{2}}$ として
       $\displaystyle{y'=\pm\frac{3}{2}x^{\frac{1}{2}}}$
       となることを確認せよ。
    1. 陰関数表示のまま両辺を微分すると $2yy'=3x^2$ となることを確認せよ。
    1. それを $y'$ について解くと
       $\displaystyle{y'=\frac{3x^2}{2y}}$
       となることを確認せよ。
    1. それに $y=\pm x^{\frac{3}{2}}$ を代入し,
       陽関数表示を微分した場合と一致することを確認せよ。

1. 微分の定義 $\displaystyle{\lim_{h\rightarrow0}\frac{f(x+h)-f(x)}{h}}$
   と $\sin x$ について
    1. 微分の定義により
       $\displaystyle{(\sin x)'=\lim_{x\rightarrow0}\frac{\sin(x+h)-\sin x}{h}}$
       となることを確認せよ。
    1. $\sin(x+h)=\sin x\cos h+\cos x\sin h$ と
       $\displaystyle{\lim_{x\rightarrow0}\frac{\sin x}{x}=1}$
       を考えて $(\sin x)'=\cos x$ となることを確認せよ。
    1. 単位円上での $\displaystyle{\frac{\pi}{2}}$ 回転を考え
       $\displaystyle{(\cos x)'}$
       $\displaystyle{=\{\sin(x+\frac{\pi}{2})'}$
       $\displaystyle{=\cos(x+\frac{\pi}{2})\}}$
       $\displaystyle{=-\sin(x)}$
       となることを確認せよ。
