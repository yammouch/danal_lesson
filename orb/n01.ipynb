```python
import numpy as np
```

```python
def symmetrize(x):
  if len(x) % 2 == 0:
    x = np.concatenate \
    ( ( x[:len(x)//2]
      , np.full(2, 0.5*x[len(x)//2])
      , x[-(len(x)//2)+1:] ) )
  return x
```

```python
def kin(ll, n):
  x = np.roll(np.arange(-(n//2), -(n//2)+n), -(n//2))
  x = -0.5*(2*np.pi/ll*(x[:, None]+x[None, :]))**2
  return x
```

```python
def pot(ll, v):
  V = np.fft.fft(v)/len(v)/ll;
  V = symmetrize(V)
  return np.diag(V)
```

```python
def cou(c, d):
  x = np.roll \
  ( np.arange(-(len(d)//2), -(len(d)//2)+len(d), dtype=np.int64)
  , -(len(d)//2) )
  t = x[None,:] - x[:,None] # t = q - p
  t_valid = (t <= len(d)//2)
  t *= t_valid
  r = x[:,None,None] + t[None,:,:] # r = s + t
  r_valid = (np.abs(r) <= len(d)//2)
  r *= r_valid
  rv \
  = np.conj(c)[:,r]     * r_valid[None,   :,:,:] \
  * d[t][None,None,:,:] * t_valid[None,None,:,:] \
  * c[:,x][:,:,None,None]
  return rv
```

```python
def exc(c, d):
  x = np.roll \
  ( np.arange(-(len(d)//2), -(len(d)//2)+len(d), dtype=np.int64)
  , -(len(d)//2) )
  s = x[None,:,None] + x[:,None,None] # s = p + t
  s_valid = (s <= len(d)//2)
  s *= s_valid
  r = x[None,None,:] - x[:,None,None] # r = q - t
  r_valid = (np.abs(r) <= len(d)//2)
  r *= r_valid
  rv \
  = np.conj(c)[:,r] * r_valid[None,:,:,:] \
  * d[x][None,:,None,None] \
  * c[:,s]          * s_valid[None,:,:,:]
  return rv
```
